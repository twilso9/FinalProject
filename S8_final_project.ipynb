```python
#allow output from every line
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
from pybaseball import statcast
```

```python
data = statcast(start_dt='2023-03-30',end_dt='2023-11-01')
```

```python
list(data)
data['description'].unique()
data['pitch_type'].unique()
```

```python
data_group = data[['player_name','game_date','pfx_x','pfx_z','pitch_type','release_speed','release_spin_rate','spin_axis','description']]
data_group
data_group_test = data_group.fillna(0)
```

```python

active = True

while active:
    while True:
        first_pitcher_input = input("Please enter the First space Last name of a pitcher: ")
        f_name_pitcher_one, l_name_pitcher_one = first_pitcher_input.split(" ")
        pitcher_name_one = l_name_pitcher_one + "," + " " + f_name_pitcher_one
        if pitcher_name_one in data['player_name'].values:
            break
        else:
            print("Name not found or spelled incorrectly.")
    while True:
        second_pitcher_input = input("Enter the first and last name of another pitcher to compare stats: ")
        f_name_pitcher_two, l_name_pitcher_two = second_pitcher_input.split(" ")
        pitcher_name_two = l_name_pitcher_two + "," + " " + f_name_pitcher_two
        if pitcher_name_two in data['player_name'].values:
            break
        else:
            print("Name not found or spelled incorrectly.")
    
    fig, axs = plt.subplots(3,2)

    sns.set_palette('tab20')
    sns.set_style("darkgrid")

    sns.scatterplot(data_group_test[(data_group_test.player_name == pitcher_name_one)],x="pfx_x",y="pfx_z",hue="pitch_type",marker="x",size=1,ax=axs[0,0])
    axs[0,0].set_title(pitcher_name_one + "\n" + "Pitch Type Location")
    sns.move_legend(axs[0,0], "upper left", bbox_to_anchor=(1,1))

    sns.scatterplot(data_group_test[(data_group_test.player_name == pitcher_name_two)],x="pfx_x",y="pfx_z",hue="pitch_type",marker="x",size=1,ax=axs[0,1])        
    axs[0,1].set_title(pitcher_name_two + "\n" + "Pitch Type Location")
    sns.move_legend(axs[0,1], "upper left", bbox_to_anchor=(1,1))

    sns.boxplot(data_group_test[(data_group_test.player_name == pitcher_name_one)], x="release_speed",y="pitch_type", ax=axs[1,0])
    axs[1,0].set_title(pitcher_name_one + "\n" + "Release Speed by Pitch Type")

    sns.boxplot(data_group_test[(data_group_test.player_name == pitcher_name_two)], x="release_speed",y="pitch_type", ax=axs[1,1])
    axs[1,1].set_title(pitcher_name_two + "\n" + "Release Speed by Pitch Type")

    sns.scatterplot(data_group_test[(data_group_test.player_name ==pitcher_name_one)],x="release_spin_rate",y="spin_axis",hue="pitch_type",size=1, ax=axs[2,0],legend=False)
    axs[2,0].set_title(pitcher_name_one + "\n" + "Spin Rate/Axis by Pitch Type")

    sns.scatterplot(data_group_test[(data_group_test.player_name == pitcher_name_two)], x="release_spin_rate",y="spin_axis",hue="pitch_type",size=1, ax=axs[2,1],legend=False)
    axs[2,1].set_title(pitcher_name_two + "\n" + "Spin Rate/Axis by Pitch Type")
    fig.subplots_adjust(wspace=.6,hspace=.65)
    fig.set_size_inches(10,10)
    fig.suptitle(pitcher_name_one + " vs " + pitcher_name_two + "\n" + "Release Speed and Pitch Movement Comparison 2023 Regular Season", fontsize=16)
    
    fig.savefig("Final_Project_Plot.pdf")
    print("PDF file saved as Final_Project_Plot.")
    
    break
```
